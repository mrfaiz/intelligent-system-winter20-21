```python
import pandas as pd
import math
import numpy as np
from sklearn.decomposition import PCA
```

```python
df_prep = pd.read_csv('combined_data_updated.csv',parse_dates=['time'],index_col='time')
df_prep.rename(columns={'temp_c_station-a':'temp_c_station_a','status_station-a':'status_station_a','rain_mm_station-a':'rain_mm_station_a','temp_c_station-b':'temp_c_station_b','status_station-b':'status_station_b','rain_mm_station-b':'rain_mm_station_b','temp_c_station-c':'temp_c_station_c','status_station-c':'status_station_c','rain_mm_station-c':'rain_mm_station_c'}, inplace = True)
```

```python
def updateDataFrameToOrdinalValues (dataFrame,column,labels):    
    max = math.ceil(dataFrame[column].max())
    min = math.floor(dataFrame[column].min())
    ranges = np.linspace(min,max,(len(labels)+1))
    print(min,max)
    print(column,ranges)
    rep_vals = []
    i=0
    while i < len(ranges)-1:
        dataFrame[column] = np.where((dataFrame[column]>ranges[i]) & (dataFrame[column]<ranges[i+1]), float(max+i), dataFrame[column])
        rep_vals.append((max+i))
        i += 1
    dataFrame[column].replace(rep_vals,labels,inplace=True)
    
```

```python
#Update data to ordinalValues

labels = ['decreased','normal','increased']
newData = df_prep.copy(deep=True)
d_types = newData.dtypes
for i in newData.columns:
    print(i)
    if (d_types[i] != 'object'):
        updateDataFrameToOrdinalValues(newData,i,labels)
        
#newData.drop(['rain_mm_station_b','rain_mm_station_a','rain_mm_station_c','temp_c_station_c','flow_m2_s'], axis=1,inplace=True)
newData
```

```python
components = ((newData.columns).to_numpy())
components = np.delete(components,-1)
components = np.delete(components,-1)
print(components)
```

```python
feature_ir = pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)
def generate_IR():
    i = 0
    while i < len(components):
        index = components[i]
        unique_labels = np.unique(newData[index].to_numpy())
        uc = 0
        IZ = np.array([])
        class_traces = pd.DataFrame(data=None, index=None, columns=labels, dtype=None, copy=False)
        while uc < len(unique_labels):
            trace = getOccurances(unique_labels[uc],newData.to_numpy(),i)
            class_traces.loc[uc] = trace 
            max_val = np.amax(trace)
            iz_score = np.sum(trace) - max_val
            IZ = np.append(IZ,iz_score)
            uc +=1
        
        class_traces['IZ'] = IZ
        IR = computeIR(class_traces)
        feature_ir[components [i]] = [IR]
        print(class_traces)
        i += 1
        
generate_IR()
```

```python
def getOccurances(label,data,feature_index):
    labels_counter_array = np.zeros([len(labels)])
    for i in data:
        if(i[feature_index] == label):
            if(i.take(-1) == 'decreased'):
                labels_counter_array[0] += 1
            elif(i.take(-1) == 'normal'):
                labels_counter_array[1] += 1
            elif(i.take(-1) == 'increased'):
                labels_counter_array[2] += 1
    return labels_counter_array
```

```python
def computeIR(data):
    total_iz = np.sum(data['IZ'].to_numpy())
    total_occurances = np.sum((data.drop(columns=['IZ'])).to_numpy())
    ir = total_iz/total_occurances
    return ir
```
