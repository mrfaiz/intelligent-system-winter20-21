```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt


col_time= "time"
col_temp_c = "temp_c"
col_status = "status"
col_rain_mm = "rain_mm"
col_level_cm = "level_cm"
col_flow_m2_s = "flow_m2_s"

statin_a_csv = "station_a.csv"
statin_b_csv = "station_b.csv"
station_main_csv = "station_main_improved.csv"
features = np.array(['h1','h2','h3','h4','h5','h6','h7','h8',
                  'h9','h10','h11','h12','h13','h14','h15','h16',
                  'h17','h18','h19','h20','h21','h22','h23'])

from sklearn.preprocessing import StandardScaler # for standardizing the Data
from sklearn.decomposition import PCA # for PCA calculation
```

```python
station_main = pd.read_csv(station_main_csv)
# y = station_main[col_level_cm].values.reshape(19, int(35063/19))
# station_main = station_main.set_index(col_time)
#station_main.index


station_main[0:25]
```

```python
water_level_data_day_basis = (pd.DataFrame(station_main.groupby(station_main.index // 23)[col_level_cm]
                          .apply(list)
                          .values
                          .tolist(), columns=features).fillna(0))

# water_level_data_day_basis.head()
```

```python
water_level_data_day_basis.shape
```

```python
from sklearn.preprocessing import StandardScaler # for standardizing the Data
from sklearn.decomposition import PCA # for PCA calculation
```

### Standardizing the data

```python
X = water_level_data_day_basis.values # getting all values as a matrix of dataframe 
sc = StandardScaler() # creating a StandardScaler object
X_std = sc.fit_transform(X) # standardizing the data
```

### Conduct PCA

```python
pca = PCA()
X_pca = pca.fit(X_std)
# pca.explained_variance_ratio_
```

### Determine the number of components 

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

### Dimensionality Reduction


```python
num_components = 4
pca = PCA(num_components)  
X_pca = pca.fit_transform(X_std) # fit and reduce dimension
```

### Finding the most important features set

```python
important_fetures_set = pd.DataFrame(pca.components_, columns = water_level_data_day_basis.columns)
important_fetures_set
```

```python
n_pcs= pca.n_components_ # get number of component
# get the index of the most important feature on EACH component
most_important = [np.abs(pca.components_[i]).argmax() for i in range(n_pcs)]
initial_feature_names = water_level_data_day_basis.columns
# get the most important feature names
most_important_features = [initial_feature_names[most_important[i]] for i in range(n_pcs)]
```

```python
most_important_features
```

```python
important_fetures_set["IZ"] = (important_fetures_set.sum(axis=1)-important_fetures_set.max(axis=1))
important_fetures_set
```

```python
IR = sum(important_fetures_set["IZ"])/4
IR
```
