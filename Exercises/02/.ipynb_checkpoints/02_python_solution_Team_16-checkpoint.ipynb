```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime

dict = {
        "10": { "begin": 20,"end":31 },
        "11": { "begin": 1, "end": 10}
}

arr = []
for month in dict:
    b = dict[month]["begin"]
    e = dict[month]["end"]
    for i in range(b, e+1):
        q = ("" if i > 9 else '0') + str(i) +'.'+month+'.2020'
        arr.append(q);

t = pd.date_range(start = pd.to_datetime(arr[0], dayfirst=True), end = pd.to_datetime(arr[len(arr)-1], dayfirst=True) + pd.DateOffset(1), freq='min')
t=t.delete(0) # cause it stats at 00:01
y = [];

for f in arr:
    pf = pd.read_csv('data/'+ f +'.csv', sep=";", encoding = "ISO-8859-1")
    for x in pd.to_numeric(pf["WSA LÜBECK"]):
        y.append(x)

frame = pd.DataFrame(data={"waterlevels": y}, index=t)
frame2 = pd.DataFrame(data={"timestamps": t, "waterlevels": y})

n = np.arange(0, len(t))
z1 = np.polyfit(n, y, 3)
p1 = np.poly1d(z1)
z2 = np.polyfit(n, y, 10)
p2 = np.poly1d(z2)

plt.plot(frame2["timestamps"], p1(n),
         frame2["timestamps"], p2(n),
         frame2["timestamps"], frame2["waterlevels"])
plt.show()
```
